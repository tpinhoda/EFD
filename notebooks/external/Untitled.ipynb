```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
#    'PSDB': [0.77, 0.61, 0.57 ,0.80, 0.76, 0.86, 0.74],
#    'PT': [0.72, 0.64, 0.55 ,0.81, 0.77, 0.84, 0.87],
    "Global Moran's Index": [0.77, 0.61, 0.57 ,0.80, 0.76, 0.86, 0.74, 0.72, 0.64, 0.55 ,0.81, 0.77, 0.84, 0.87],
    "Party": ['PSDB', 'PSDB', 'PSDB', 'PSDB', 'PSDB', 'PSDB', 'PSDB', 'PT', 'PT', 'PT', 'PT', 'PT', 'PT', 'PT'],
    'Years': [1994, 1998, 2002, 2006, 2010, 2014, 2018, 1994, 1998, 2002, 2006, 2010, 2014, 2018] 
}
df = pd.DataFrame(data)
df = df[::-1]
df
```

```python
fig, ax = plt.subplots(figsize=(15,5))
sns.lineplot(ax=ax,
              x = "Years", 
              y = "Global Moran's Index", 
              hue='Party',
              style="Party",
              palette = "Set1",
              marker="o",
              data=df)
ax.set_xlabel("Election Years",fontsize=20)
ax.set_ylabel("Global Moran's Index",fontsize=20)
ax.tick_params(labelsize=15)
alig = {'PSDB': ['right', 'bottom', -0.015],
        'PT': ['left', 'top', 0.015]}
ax.set(xticks=df.Years)
for idx, item in df.groupby('Party'):
    #item[1] is a grouped data frame
    for x,y in item[['Years',"Global Moran's Index"]].values:
        ax.text(x+alig[idx][2],y+alig[idx][2],y,color='black', 
                ha=alig[idx][0],
                va= alig[idx][1],
                fontsize='large',
                weight = 'demi',
                stretch='ultra-condensed'
               )
plt.legend(fontsize='x-large', title_fontsize='40')
plt.savefig("global_morans_idex.png")
```
